```python
import matplotlib.pyplot as plt
import numpy as np

# polynomial fit of degree 5
n = 5
# loading x-y data set
with open('/Users/cs0823/Documents/ECS230_HW4/data/data.dat') as dataset:
    next(dataset) #ignore the first line with '10'
    datapair = dataset.readlines()
x = []
y = []
for i in datapair:
    i = i.split()
    x.append(float(i[0]))
    y.append(float(i[1]))

# load fitting data
with open('/Users/cs0823/Documents/ECS230_HW4/data/degree5.dat') as file:
    data = file.readlines()

fitting_data = data[-1]
fitting_data = fitting_data.split()
fitting_data = [float(i) for i in fitting_data]
fitting_data.reverse()
del fitting_data[0:n-1]

# visualization
plt.figure()
degree5_parameters = np.poly1d(fitting_data)
plt.plot(x, y, 'bo')
xl = np.linspace(0,5.3)

plt.plot(xl,degree5_parameters(xl))
plt.xlabel('x')
plt.ylabel('y')
plt.title("polynomial fit of degree 5")
yl = plt.gca()
yl.set_ylim([0,10])
plt.show()
```
